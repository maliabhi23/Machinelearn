```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python

# Step 2: Load the data
data = pd.read_csv("Mall_Customers.csv")

# Step 3: Preprocess the data
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.iloc[:, 2:])  # Scaling only numerical features

```

```python
kmeans = KMeans(n_clusters=5, random_state=42)  # Specify the number of clusters
kmeans.fit(scaled_data)

```

```python
data['Cluster'] = kmeans.labels_
```

```python
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], c=data['Cluster'], cmap='viridis')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('K-Means Clustering')
plt.show()
```

```python
print("Cluster Centers:")
print(scaler.inverse_transform(kmeans.cluster_centers_))
```

```python

# Load your data
data = pd.read_csv("segmentation datam.csv")
```

```python

# Select features for clustering
X = data[['Sex', 'Marital status', 'Age', 'Education', 'Income', 'Occupation', 'Settlement size']]
```

```python

# Choose the number of clusters (k)
k = 3
```

```python

# Initialize KMeans object
kmeans = KMeans(n_clusters=k)
```

```python

# Fit KMeans clustering model to the data
kmeans.fit(X)
```

```python

# Get cluster labels for each data point
cluster_labels = kmeans.labels_

```

```python
# Add cluster labels to the original DataFrame
data['Cluster'] = cluster_labels

```

```python
# View the resulting clusters
print(data['Cluster'].value_counts())
```
